```python
def recuperardados(fnome):
    dados=[]
    fdados=open(fnome)
    for linha in fdados:
        termos=linha[:-1].split(",")
        dados.append(termos)    
    fdados.close()
    return dados

recuperardados("diabetes_prediction_dataset.csv")
```

```python
dados=recuperardados("diabetes_prediction_dataset.csv")
def sexoDist(lista):
    m=0
    f=0
    for x in lista:
        if x[-1]== '1':
            if x[0]=='Male':
                m=m+1
            else:
                f=f+1
    distSexo=[("F",f),("M",m)]
    return distSexo

sexoDist(dados)
```

```python
def escaloesEtarios(lista):
    idadeMax=0
    for i in lista:
        if float(i[1])>idadeMax:
            idadeMax = float(i[1])

    while idadeMax%5 != 0:
        idadeMax += 1

    escaloes = [[(0,10),0], [(11,24),0]]+[[(x, x+4), 0] for x in range(25, int(idadeMax), 5)]

    for i in lista:
        if i[-1] == "1":
            idade= float(i[1])
            if idade<=10:
                if escaloes[0][0][0]<= idade and idade<=escaloes[0][0][1]:
                    escaloes[0][1] += 1

            if idade<25 and idade>10:
                if escaloes[1][0][0]<= idade and idade<=escaloes[1][0][1]:
                    escaloes[1][1] += 1

            if idade>=25:
                for escalao in escaloes:
                    if idade >= escalao[0][0] and idade <= escalao[0][1]:
                        escalao[1] += 1

    return escaloes

dados=recuperardados("diabetes_prediction_dataset.csv")

escaloesEtarios(dados[1:])
```

```python
def glicose(lista):
    glicoseMax=0
    for i in lista:
        if int(i[-2])>glicoseMax:
            glicoseMax = int(i[-2])

    while glicoseMax%10 != 0:
        glicoseMax += 1
    
    glicoseMin=glicoseMax

    for i in lista:
        if int(i[-2])<glicoseMin:
            glicoseMin=int(i[-2])

    while glicoseMin%10 !=0:
        glicoseMin -= 1
    
    escaloes = [[(x, x+9), 0] for x in range(glicoseMin, glicoseMax, 10)]

    for i in lista:
        if i[-1] == "1":
            glicose= int(i[-2])
            for escalao in escaloes:
                    if glicose >= escalao[0][0] and glicose <= escalao[0][1]:
                        escalao[1] += 1

    return escaloes

glicose(dados[1:])
```

```python
def tabela(dados,critrio):
    if critrio==sexoDist:
        print(" Sexo   | n de pessoas doentes")
    if critrio==escaloesEtarios:
        print("Distribuio etria   | n de pessoas doentes")
    if critrio==glicose:
        print("Diabetes | n de doentes")
    for i in critrio(dados):
        print(i[0], "|", i[1])
```

```python
def menu():
    print("""Bem-vindo as distribuies relativamente  diabetes 
Menu:
    (1) Consultar distribuio por sexo
    (2) Consultar distribuio por faixa etria 
    (3) Consultar distribuio por nvel de glicose no sangue
    (0) Sair """)
    return


    menu()
function=int(input("O que pretende fazer?"))

while function !=0:
    if function == 1:
        tabela(dados[1:],sexoDist)

    if function == 2:
        tabela(dados[1:],escaloesEtarios)

    if function == 3:
       tabela(dados[1:],glicose)
    

        
    function = int(input("O que pretende consultar de seguida?"))
print("")
print("Saiu do programa das distribuies")
```
